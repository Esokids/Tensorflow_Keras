```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras import layers
```

```python
N = 100
X = np.random.rand(N, 1)
Y = np.sin(2*np.pi*X) + 0.4*np.random.rand(N, 1)
plt.plot(X, Y, '.')
plt.show()
```

```python
# inputs 1 -- 100 -- 100 -- outputs 1
model = keras.Sequential()
model.add(layers.Dense(100, activation='relu')) # you can assign or not assign input shape
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(1))
model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.1), loss=keras.losses.MeanSquaredError())
```

```python
model.fit(X, Y, epochs=500) # batch size befault is 32 -- 1 epoch train several times
```

```python
Z = model.predict(X)
plt.plot(X, Y, '.')
plt.plot(X, Z, '.r')
plt.show()
```
